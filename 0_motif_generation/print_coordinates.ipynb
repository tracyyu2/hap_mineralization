```python
def parse_pdb(pdb_file):
    atom_coords = []
    with open(pdb_file, 'r') as f:
        for line in f:
            if line.startswith('ATOM') or line.startswith('HETATM'):
                # Example format: ATOM   452  O2   CO3   126       4.474   7.526   7.491  1.00  7.97           O  
                # Extracting coordinates
                x = float(line[30:38])
                y = float(line[38:46])
                z = float(line[46:54])
                atom_coords.append((x, y, z))
    
    return atom_coords

def write_coordinates_to_file(coordinates, output_file):
    with open(output_file, 'w') as f:
        for coord in coordinates:
            f.write(f"X: {coord[0]}, Y: {coord[1]}, Z: {coord[2]}\n")

# Example usage:
pdb_file = '/home/tracyy2/HA_binder_design/example_glu.pdb'
output_file = '/home/tracyy2/HA_binder_design/coordinates_output.txt'

# Parse PDB file and get coordinates
coordinates = parse_pdb(pdb_file)

# Write coordinates to text file
write_coordinates_to_file(coordinates, output_file)

print(f"Coordinates extracted from {pdb_file} have been saved to {output_file}.")

```
